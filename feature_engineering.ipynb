```python
import pandas as pd
materials = pd.read_csv("data\material.csv")
products = pd.read_csv("data\product.csv")
```

```python
# Feature 1: CO₂ Impact Index
materials["co2_impact_index"] = (
    materials["biodegradability_score"] / materials["co2_emission_score"]
)
```

```python
avg_weight = products["weight_capacity_score"].mean()
```

```python
# Feature 2: Estimated Packaging Cost
materials["estimated_cost"] = (
    (0.5 * materials["strength_score"]) +
    (0.3 * materials["co2_emission_score"]) +
    (0.2 * (1 - materials["biodegradability_score"]))
)
```

```python
# Feature 3: Cost Efficiency Index
materials["cost_efficiency_index"] = (
    materials["strength_score"] / materials["estimated_cost"]
)
```

```python
# Feature 4: Material Suitability Score
# (Material-only, no product dependency)
# ------------------------------------
materials["material_suitability_score"] = (
    (0.4 * materials["strength_score"]) +
    (0.3 * materials["biodegradability_score"]) -
    (0.3 * materials["co2_emission_score"])
)
```

```python
import os

# Define the directory path
output_dir = "/mnt/data"

# Create the directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Save final dataset
materials.to_csv(
    os.path.join(output_dir, "material_feature_engineered.csv"),
    index=False
)

print("✅ Feature engineering completed correctly (no missing columns)")
```
