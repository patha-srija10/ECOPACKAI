```python
import pandas as pd
# Load dataset
df = pd.read_csv("data\material.csv")
```

```python
# 1. Clean column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
# 2. Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# 3. Handle missing values
# Drop rows where material name is missing
df.dropna(subset=["material_name"], inplace=True)

# Fill numeric missing values with column mean
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill text missing values with 'unknown'
text_cols = df.select_dtypes(include=["object"]).columns
df[text_cols] = df[text_cols].fillna("unknown")
```

```python
import pandas as pd
# 4. Convert data types
# Diagnostic: Print columns to verify actual names after cleaning
print(df.columns)
df["strength_score"] = df["strength_score"].astype(float)
df["co2_emission_score"] = df["co2_emission_score"].astype(float)
df["biodegradability_score"] = df["biodegradability_score"].astype(float)
```

```python
# 5. Standardize text values
df["material_name"] = df["material_name"].str.title()
```

```python
import os

base_dir = os.path.abspath(os.path.join(os.getcwd(), "data"))
os.makedirs(base_dir, exist_ok=True)

output_path = os.path.join(base_dir, "materials_cleaned.csv")

df.to_csv(output_path, index=False)
print("materials_cleaned.csv REALLY created at:", output_path)
```
