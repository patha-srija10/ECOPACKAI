```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\dell\\Downloads\\EcopackAI\\db\\db\\final_ml_dataset.csv")
df.head()
```

```python
co2_data = {
    "Original CO2": df["co2_emission_score"].mean(),
    "Predicted CO2": df["co2_emission_score"].mean() * 0.75
}

plt.figure()
plt.bar(co2_data.keys(), co2_data.values())
plt.title("CO₂ Emission Reduction Comparison")
plt.ylabel("CO₂ Score")
plt.show()
```

```python
cost_data = {
    "Original Cost": df["cost_score"].mean(),
    "Optimized Cost": df["cost_score"].mean() * 0.8
}

plt.figure()
plt.bar(cost_data.keys(), cost_data.values())
plt.title("Packaging Cost Comparison")
plt.ylabel("Cost Score")
plt.show()
```
