```python
df = pd.read_csv("final_ml_dataset.csv")

X = df[
    [
        "weight_capacity_score",
        "product_strength_req",
        "barrier_score",
        "reuse_potential_score",
        "material_strength",
        "biodegradability",
        "recyclability_percent_y"
    ]
]

print(X.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train_cost, X_test_cost, y_train_cost, y_test_cost = train_test_split(
    X_scaled, y_cost, test_size=0.2, random_state=42
)

X_train_co2, X_test_co2, y_train_co2, y_test_co2 = train_test_split(
    X_scaled, y_co2, test_size=0.2, random_state=42
)
```

```python
#STEP 3
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
```

```python
#split data for cost model
X_train_cost, X_test_cost, y_train_cost, y_test_cost = train_test_split(
    X_scaled,
    y_cost,
    test_size=0.2,
    random_state=42
)
```

```python
#split data for co2 model
X_train_co2, X_test_co2, y_train_co2, y_test_co2 = train_test_split(
    X_scaled,
    y_co2,
    test_size=0.2,
    random_state=42
)
```

```python
#Train Random Forest model
cost_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

cost_model.fit(X_train_cost, y_train_cost)
```

```python
#predict cost
y_pred_cost = cost_model.predict(X_test_cost)
```

```python
#evaluate cost model
rmse_cost = np.sqrt(mean_squared_error(y_test_cost, y_pred_cost))
mae_cost = mean_absolute_error(y_test_cost, y_pred_cost)
r2_cost = r2_score(y_test_cost, y_pred_cost)

print("COST MODEL PERFORMANCE")
print("RMSE:", rmse_cost)
print("MAE :", mae_cost)
print("R²  :", r2_cost)
```

```python
#co2 footprint prediction - 2nd model
#train co2 model
co2_model = GradientBoostingRegressor(
    n_estimators=100,
    random_state=42
)

co2_model.fit(X_train_co2, y_train_co2)
```

```python
#predict co2
y_pred_co2 = co2_model.predict(X_test_co2)
```

```python
#evaluate co2 model
rmse_co2 = np.sqrt(mean_squared_error(y_test_co2, y_pred_co2))
mae_co2 = mean_absolute_error(y_test_co2, y_pred_co2)
r2_co2 = r2_score(y_test_co2, y_pred_co2)

print("CO₂ MODEL PERFORMANCE")
print("RMSE:", rmse_co2)
print("MAE :", mae_co2)
print("R²  :", r2_co2)
```

```python
#compare models in table
import pandas as pd

results = pd.DataFrame({
    "Model": ["Cost Prediction", "CO₂ Prediction"],
    "RMSE": [rmse_cost, rmse_co2],
    "MAE": [mae_cost, mae_co2],
    "R²": [r2_cost, r2_co2]
})

results
```

```python
df["predicted_cost"] = cost_model.predict(X_scaled)
df["predicted_co2"] = co2_model.predict(X_scaled)
```

```python
df[["predicted_cost", "predicted_co2"]].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[["cost_norm", "co2_norm"]] = scaler.fit_transform(
    df[["predicted_cost", "predicted_co2"]]
)
```

```python
df[["strength_gap_norm", "weight_gap_norm"]] = scaler.fit_transform(
    df[["strength_gap", "weight_gap"]]
)
```

```python
df["final_score"] = (
    (1 - df["cost_norm"]) +
    (1 - df["co2_norm"]) +
    df["strength_gap_norm"] +
    df["weight_gap_norm"]
)
```

```python
df[["final_score"]].head()
```

```python
df["rank"] = df.groupby("product_id")["final_score"] \
               .rank(method="dense", ascending=False)
```

```python
top_3_recommendations = df[df["rank"] <= 3] \
    .sort_values(["product_id", "rank"])
```

```python
top_3_recommendations[
    [
        "product_id",
        "product_name",
        "material_name",
        "predicted_cost",
        "predicted_co2",
        "final_score",
        "rank"
    ]
].head(10)
```

```python
top_3_recommendations.to_csv(
    "material_recommendations.csv",
    index=False
)
```
